```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
np.random.seed(0)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value': np.random.randint(1, 100, size=100),
    'Value2': np.random.normal(0, 1, 100)
})
plt.figure(figsize=(8, 6))
plt.scatter(data['Value'], data['Value2'])
plt.xlabel('Value')
plt.ylabel('Value2')
plt.title('Scatter Plot')
plt.show()


```

```python
plt.figure(figsize=(8, 6))
data.groupby('Category')['Value'].mean().plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Average Value')
plt.title('Bar Chart of Average Value per Category')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Value'], bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Value')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=data)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot of Value per Category')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data['Value'].cumsum())
plt.xlabel('Index')
plt.ylabel('Cumulative Sum of Value')
plt.title('Line Plot of Cumulative Value')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Category Distribution')
plt.ylabel('')
plt.show()
```
